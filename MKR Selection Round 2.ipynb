# IMORTING AND CLEANING DATA

##### I am here importing the data and cleaning the empty and absurd values out of the data to make it ready for analysis.

```python
from pandas import *
```

```python
df = read_csv(r'C:\Users\harsi\OneDrive\Documents\MENTAL HEALTH DATASET - MENTAL HEALTH DATASET.csv')
df
```

```python
df['Gender'] = df['Gender'].replace(dict.fromkeys(['M','Male ','Make','Mail','Mal','Male-ish','male','m','Male (CIS)','Malr','Man ','cis male','maile','msle','Cis Male','Cis Man','Man'], 'Male'))
df['Gender'] = df['Gender'].replace(dict.fromkeys(['Cis Female','F','Femake','Female ','Female (cis)','Woman','cis-female/femme','f','femail','female','woman'], 'Female'))
df['Gender'] = df['Gender'].replace(dict.fromkeys(['Female (trans)','Trans woman','Trans-female','Genderqueer','queer','queer/she/they','non-binary','Agender','fluid'], 'Trans'))
```

```python
df1 = df.loc[df['Gender'].isin(['Male', 'Female', 'Trans'])]
df2 = df1.loc[(df['Age'].isin(range(0,60)))]
df2
```

```python
df2[['state', 'self_employed']] = df2[['state', 'self_employed']].fillna('Not Known')
df2['work_interfere'] = df2['work_interfere'].fillna('No Work')
df2['comments'] = df2['comments'].fillna('No Comments')
df2
```
